```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
import itertools
np.random.seed(0)
Learning_rate = 0.01
```

```python
learning_rate = 0.01
iris = load_iris()
X = iris.data[:,:2] #First two features of iris dataset
Y = iris.target
#print(X)


Y = (Y!=0)*1 #create Y as either 0 or 1


#standardize
mean =np.mean(X,axis=0)
std=np.std(X,axis=0,ddof=1)
X = (X-mean)/std



# logistic regression from scratch

#sigmoid fucntion
def sigmoid(x):
    return 1 / (1+np.exp(-x))

#cost function
def costfunc(theta,X,Y):
    m = Y.size
    dummy = sigmoid(X@theta)
    cost = -1 * np.sum(Y*np.log(dummy) + (1-Y) * np.log(1-dummy)) / m
    return cost

#gradient descent function uses cost function and sigmoid function. Runs until 10,000 iterations or until cost is less than 2^-23
def gradient_descent(X,Y,thetas):
    m,num_features = X.shape[:2]
    #print(m)
    #print(num_features)
    biasset = np.ones((X.shape[0],1))
    X = np.concatenate((biasset,X),axis=1)
    Y = np.array(Y)[:,np.newaxis]
    
    i=0
    while (i<10000):
        i+=1
        exp = X@thetas
        sigmoidval = sigmoid(exp)
        gradient = 1/m * (X.T@(sigmoidval-Y))
        thetas -= gradient*learning_rate
        cost = costfunc(thetas,X,Y)
        if (cost<=2**-23):
            break
        
    return thetas

#initialize random thetas
Theta= 2*np.random.random_sample((3,1))-1 
thetas = gradient_descent(X,Y,Theta)

#print('Thetas',thetas)  #uncomment to print all thetas
```

```python
colors = itertools.cycle(["r", "b"])
#plt.scatter(X[:,0],X[:,1],color = next(colors))
plt.scatter(X[:,0],X[:,1],c=Y)
x1_min= X[:,0].min()
x1_max = X[:,0].max()
x2_min = X[:,1].min(), 
x2_max = X[:,1].max()

x1_bdy = np.linspace(x1_min,x1_max,10000)
x2_bdy = (0.6-thetas[0]-thetas[1] * x1_bdy) / thetas[2]
plt.plot(x1_bdy,x2_bdy,label='Logistic Regression Boundary for Iris')
plt.title('Logistic Regression Boundary for Iris')
plt.show()
 
#check using sklearn functions, purely for validation purposes
lgr = LogisticRegression(solver ='lbfgs',max_iter=10000)
l =lgr.fit(X,Y)
print(l)
print (lgr.intercept_)
print(lgr.coef_)
```
